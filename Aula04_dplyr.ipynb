```r
# Importar um arquivo com informações do PIB
# Verifique se seu arquivo tem virgula ou ponto e virgula como separador.
# Baixe o arquivo em https://drive.google.com/open?id=1L6EHrw7DqNkZVGj3GRBPM_DHyyLmH1xc
pib <- read.csv(file="./dados/PIB/PIB_2010-2015.csv", header=TRUE, sep=",", stringsAsFactors=FALSE)
head(pib)
```

```r
#Seleção do PIB de 2010. Esse dataset tem dados de 2010 a 2015
# Filtra (retorna) somente as linhas
pib2010 <- pib %>% filter(Ano == 2010)
head(pib2010)
```

```r
#calculo da média do PIB de 2010 por UF
media_uf <- pib %>% 
    filter(Ano == 2010) %>%
    group_by(UF) %>% 
        summarise(media=mean(PIB, na.rm=TRUE))
media_uf
```

```r
#Seleção dos dados dos estados de São Paulo, Rio de Janeiro e Minas Gerais
estados <- c("São Paulo", "Rio de Janeiro", "Minas Gerais")
alguns_estados <- pib %>%
    select(UF,PIB,Ano) %>%
    filter(UF %in% estados & Ano == 2010)
alguns_estados
```

```r
#Calculo do PIB médio dos estados de São Paulo, Rio de Janeiro e Minas Gerais
estados <- c("São Paulo", "Rio de Janeiro", "Minas Gerais")
media_algumas_uf <- pib %>%
    filter(UF %in% estados & Ano == 2010) %>% 
        group_by(UF) %>% 
            summarise(media=mean(PIB, na.rm=TRUE), soma=sum(PIB))
media_algumas_uf
```
